```python
# Create Color Pallete
import matplotlib.pyplot as plt
import numpy as np

# Define pastel color palette
pastel_colors = [
    '#fbb4ae', # Light Pink
    '#b3cde3', # Light Blue
    '#ccebc5', # Light Green
    '#decbe4', # Light Purple
    '#fed9a6', # Light Orange
    '#ffffcc', # Light Yellow
    '#e5d8bd', # Light Beige
    '#f2f2f2', # Light Gray
]

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 2))

# Plot colors with labels
for i, color in enumerate(pastel_colors):
    ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
    ax.text(i + 0.5, 0.5, color, color='black', ha='center', va='center', fontsize=12)

# Set limits and hide axes
ax.set_xlim(0, len(pastel_colors))
ax.set_ylim(0, 1)
ax.axis('off')

# Save image
plt.savefig('pastel_colors_with_hex.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
